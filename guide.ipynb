# Siglas e seus significados 

P: Quantidade de elementos

V: Quantidade variáveis

C: Quantidade de Clusters  

i: Elemento

j: Variável (feature)

k: cluster



Elementos: $E = \{e_1,\dots , e_p\},   e_i (1 \leq i \leq  P)$

Variáveis:  $X_i = \{x_{i1}, \dots  , x_{iV}\}, x_{ij}(1 \leq j \leq  V)$

Grau de pertinência: $U = (u_1, \dots , u_p) = (u_{ik})_{1 \leq i \leq  P, 1 \leq k \leq  C}$

Protótipos: $G = (g_1, \dots , g_C) = (g_{kj})_{1 \leq k \leq  C, 1 \leq j \leq  V}$

O componente $g_k = (g_{k1}, \dots, g_{kV})$ é o protótipo (representante) do cluster fuzzy $k$. Há apenas um protótipo por cluster 

$T_u$: Grau de fuzzificação

$T_v$: Relevância das variáveis
