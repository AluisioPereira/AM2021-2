# Criação dos relatórios exigidos na questão 2

* Basta executar diretamente todos os módulos listados anteriormente para obter os relatórios, que são sempre  gerados na pasta reports.
* Todos os scores obtidos em cada split de treinamento estão disponíveis no arquivo MODELO_splits_scores.csv. Esses relatórios são meramente para log e não são exigidos pelo professor. Este arquivo contém apenas os melhores resultados da validação, incluindo 
* O melhor parâmetro obtido, quando existir, e o melhor resultado estão presentes em MODELO_best_result.csv
* O desempenho final do modelo, testado com a base de teste, está presente no arquivo MODELO_classification_report.txt
* O desempenho geral do modelo, a estimativa pontual e o intervalo de confiança estão presentes no arquivo MODELO_estimativa_pontual_intervalo.txt. 
* Ao final da execução de cada módulo, é gerado um arquivo chamado MODELO_scores_for_friedman_test.csv, contendo 300 execuções a serem usadas para comparar os modelos usando o friedman test.
* Freadman test é executando executando-se o módulo src/questao2/friedman_test.py.
* O pós-teste de Nemenyi é obtido executando o script reports/friedman_test_pos_test.r (gerado no passo anterior). É um script em R que pode ser executado online e foi escolhido por ter um output mais amigável que as bibliotecas python encontrados. O resultado final do relatório gerado por esse script é a resposta do pós teste  de Nemenyi.
