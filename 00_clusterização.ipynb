```python
import pickle
from tqdm import tqdm 
import src.dataset as dataset

from src.questao1 import executar_treinamentos
```

```python
QTD_TREINOS = 1
```

```python
def treinar_com_dataset(dataset, qtd_treinos = QTD_TREINOS, seed=12022):
  resultado = executar_treinamentos(dataset, seed=seed)
  melhor = resultado
  comp = lambda res: res["custo"]

  for t in tqdm(range(1, qtd_treinos)):
    resultado = executar_treinamentos(dataset, seed=seed+t)
    melhor = min(melhor, resultado, key = comp)
  
  return melhor

def exportar_resultado(resultado, caminho):
      with open(caminho, "wb") as f:
        pickle.dump(resultado, f)
```

# Clusterizando

## RGB

```python
melhor_rgb = treinar_com_dataset(dataset.RGB_TEST)
exportar_resultado(melhor_rgb, "data/melhor_resultado_rgb.pickle")
print("\nMelhor custo com RGB: ", melhor_rgb['custo'])
```

## SHAPE

```python
melhor_shape = treinar_com_dataset(dataset.SHAPE_TEST)
exportar_resultado(melhor_shape, "data/melhor_resultado_shape.pickle")
print("\nMelhor custo com SHAPE: ", melhor_shape['custo'])
```

## RGB_SHAPE

```python
melhor_rgb_shape = treinar_com_dataset(dataset.SHAPE_RGB_TEST)
exportar_resultado(melhor_rgb_shape, "data/melhor_resultado_rgb_shape.pickle")
print("\nMelhor custo com SHAPE_RGB: ", melhor_rgb_shape['custo'])
```
